# Testing parameter tuning general view

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy
```

```python
init_temp = 15
ratio = 1.00020
avg_val_diff = 5
max_iter = 10000
```

# Testing average difference

avg diff = 4

```python
avg_val_tuning_value = []    
for avg_val_diff in range(1, 10):
    
    
    value_diff_per_current_temp = []
    for step in range(0, max_iter):

        current_temperature = init_temp/(ratio**step)
        value_diff_per_current_temp.append((-avg_val_diff)/current_temperature)

    exp_y = numpy.exp(value_diff_per_current_temp)
    
    avg_val_tuning_value.append(exp_y)
```

```python
df = pd.DataFrame(avg_val_tuning_value, index=[(i) for i in range(1, 10)]).T

plt.figure(figsize=(13, 13))
plt.plot(df)
plt.legend(df.columns);
```

# Testing ratio difference

```python
test_range = (0, 10)
skip = 10
```

```python
ratio_tuning_value = []    
for i in range(test_range[0], test_range[1]):
    
    test_ratio = (ratio * 100000 + i * skip ) / 100000
    
    value_diff_per_current_temp = []
    for step in range(0, max_iter):

        current_temperature = init_temp/(test_ratio**step)
        value_diff_per_current_temp.append((-avg_val_diff)/current_temperature)

    exp_y = numpy.exp(value_diff_per_current_temp)
    
    ratio_tuning_value.append(exp_y)
```

```python
df = pd.DataFrame(ratio_tuning_value, index=[ (ratio * 100000 + i * skip ) / 100000 for i in range(test_range[0], test_range[1])]).T

plt.figure(figsize=(13, 13))
plt.plot(df)
plt.legend(df.columns);
```

# Testing init_temp difference

```python
test_range = (0, 10)
skip = 10
start_offset = 10
```

```python
init_temp_tuning_value = []    
for i in range(test_range[0], test_range[1]):
    
    test_temp = start_offset + i * skip
    
    value_diff_per_current_temp = []
    for step in range(0, max_iter):

        current_temperature = test_temp/(ratio**step)
        value_diff_per_current_temp.append((-avg_val_diff)/current_temperature)

    exp_y = numpy.exp(value_diff_per_current_temp)
    
    init_temp_tuning_value.append(exp_y)
```

```python
df = pd.DataFrame(init_temp_tuning_value, index=[(start_offset + i * skip) for i in range(test_range[0], test_range[1])]).T

plt.figure(figsize=(13, 13))
plt.plot(df)
plt.legend(df.columns);
```
